```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlt
labels  =  "TDP","YSRCP","JSP+","OTHERS" 
sizes  =  [83,63,18,7]
colors=["yellow","green","red","blue"] 
explode  =  (0.0,0.0,0.1,0.0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.2f%%',  shadow=True)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlt
labels  =  "TDP","YSRCP","JSP+","OTHERS" 
sizes  =  [79,70,18,7]
colors=["yellow","green","red","blue"] 
explode  =  (0,0.0,0.0,0.0)
patches,text  =  plt.pie(sizes,explode=explode,colors=colors,  shadow=True,  startangle=90)
plt.legend(patches, labels, loc='center left')
plt.tight_layout() 
plt.axis('equal') 
plt.show()

```

```python
import matplotlib.pyplot as plt 
import matplotlib.mlab as mlt 
import pandas as pd
df  =  pd.read_excel("D:\santo.xlsx") 
name = df['name']
marks = df['marks']
plt.plot(marks,name,color="red",alpha=.8)
plt.xlabel("names",color="green",alpha=1) 
plt.ylabel("Marks",  color="blue",alpha=1) 
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
objects  =  ('TDP',  'YSRCP',  'JSP',  'CPI/CPM',  'CONG',  'OTHERS')
y_pos  =  np.arange(len(objects)) 
performance  =  [77,59,16,10,3,12]
plt.bar(y_pos,  performance,  align='center',  alpha=1,  color=("yellow","green","orange","red"))
plt.xticks(y_pos, objects)
plt.ylabel('SEATS') 
plt.title('PARTIES')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
N = 50
colors = np.random.rand(N)
x = np.random.rand(N)
y = np.random.rand(N)
area = np.pi*(15*np.random.rand(N)**2)

# Plot
plt.scatter(x, y, s=area, c = colors,alpha=.8)
plt.title('population in india')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
